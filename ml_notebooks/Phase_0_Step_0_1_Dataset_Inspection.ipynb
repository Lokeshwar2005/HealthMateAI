# Phase 0, Step 0.1: Dataset Acquisition & Inspection

**Objective:** Load the three UCI datasets (Heart Disease, Diabetes, CKD) and perform initial inspection.

**Outputs:**
- Dataset shape, dtypes, missing value counts
- Target variable distributions
- Summary statistics (numeric columns)
- Feature count and type breakdown

```python
# Import libraries
import pandas as pd
import numpy as np
import os
from ucimlrepo import fetch_ucirepo

print("Libraries imported successfully")
```

## 1. Load Heart Disease Dataset (UCI ID 45)

```python
# Fetch Heart Disease dataset
heart_disease = fetch_ucirepo(id=45)
X_hd = heart_disease.data.features
y_hd = heart_disease.data.targets

print("\n=== HEART DISEASE DATASET ===")
print(f"Features shape: {X_hd.shape}")
print(f"Target shape: {y_hd.shape}")
print(f"\nFeature columns: {list(X_hd.columns)}")
print(f"\nTarget name: {y_hd.columns[0] if len(y_hd.columns) > 0 else 'N/A'}")
```

```python
# Heart Disease: data types and missing values
print("\n--- Data Types ---")
print(X_hd.dtypes)
print("\n--- Missing Values (Features) ---")
print(X_hd.isnull().sum())
print("\n--- Missing Values (Target) ---")
print(y_hd.isnull().sum())
```

```python
# Heart Disease: target distribution
print("\n--- Target Distribution ---")
print(y_hd[y_hd.columns[0]].value_counts())
print(f"\nTarget proportions:")
print(y_hd[y_hd.columns[0]].value_counts(normalize=True))
```

```python
# Heart Disease: summary statistics
print("\n--- Summary Statistics (numeric features) ---")
print(X_hd.describe())
```

## 2. Load Diabetes Dataset (UCI ID 891)

```python
# Fetch Diabetes dataset
diabetes = fetch_ucirepo(id=891)
X_dm = diabetes.data.features
y_dm = diabetes.data.targets

print("\n=== DIABETES DATASET ===")
print(f"Features shape: {X_dm.shape}")
print(f"Target shape: {y_dm.shape}")
print(f"\nFeature columns: {list(X_dm.columns)}")
print(f"\nTarget name: {y_dm.columns[0] if len(y_dm.columns) > 0 else 'N/A'}")
```

```python
# Diabetes: data types and missing values
print("\n--- Data Types ---")
print(X_dm.dtypes)
print("\n--- Missing Values (Features) ---")
print(X_dm.isnull().sum())
print("\n--- Missing Values (Target) ---")
print(y_dm.isnull().sum())
```

```python
# Diabetes: target distribution
print("\n--- Target Distribution ---")
print(y_dm[y_dm.columns[0]].value_counts())
print(f"\nTarget proportions:")
print(y_dm[y_dm.columns[0]].value_counts(normalize=True))
```

```python
# Diabetes: summary statistics
print("\n--- Summary Statistics (numeric features) ---")
print(X_dm.describe())
```

## 3. Load Chronic Kidney Disease Dataset (UCI ID 336)

```python
# Fetch CKD dataset
ckd = fetch_ucirepo(id=336)
X_ckd = ckd.data.features
y_ckd = ckd.data.targets

print("\n=== CHRONIC KIDNEY DISEASE DATASET ===")
print(f"Features shape: {X_ckd.shape}")
print(f"Target shape: {y_ckd.shape}")
print(f"\nFeature columns: {list(X_ckd.columns)}")
print(f"\nTarget name: {y_ckd.columns[0] if len(y_ckd.columns) > 0 else 'N/A'}")
```

```python
# CKD: data types and missing values
print("\n--- Data Types ---")
print(X_ckd.dtypes)
print("\n--- Missing Values (Features) ---")
print(X_ckd.isnull().sum())
print("\n--- Missing Values (Target) ---")
print(y_ckd.isnull().sum())
```

```python
# CKD: target distribution
print("\n--- Target Distribution ---")
print(y_ckd[y_ckd.columns[0]].value_counts())
print(f"\nTarget proportions:")
print(y_ckd[y_ckd.columns[0]].value_counts(normalize=True))
```

```python
# CKD: summary statistics
print("\n--- Summary Statistics (numeric features) ---")
print(X_ckd.describe())
```

## 4. Summary Report

```python
print("\n" + "="*80)
print("PHASE 0 STEP 0.1 INSPECTION SUMMARY")
print("="*80)

summary_table = pd.DataFrame({
    'Dataset': ['Heart Disease', 'Diabetes', 'CKD'],
    'Samples': [X_hd.shape[0], X_dm.shape[0], X_ckd.shape[0]],
    'Features': [X_hd.shape[1], X_dm.shape[1], X_ckd.shape[1]],
    'Missing (%)': [
        f"{X_hd.isnull().sum().sum() / (X_hd.shape[0] * X_hd.shape[1]) * 100:.1f}",
        f"{X_dm.isnull().sum().sum() / (X_dm.shape[0] * X_dm.shape[1]) * 100:.1f}",
        f"{X_ckd.isnull().sum().sum() / (X_ckd.shape[0] * X_ckd.shape[1]) * 100:.1f}"
    ],
    'Target Positive (%)': [
        f"{(y_hd[y_hd.columns[0]] > 0).sum() / len(y_hd) * 100:.1f}",
        f"{(y_dm[y_dm.columns[0]] == 1).sum() / len(y_dm) * 100:.1f}",
        f"{(y_ckd[y_ckd.columns[0]] == 'ckd').sum() / len(y_ckd) * 100:.1f}"
    ]
})

print("\n" + summary_table.to_string(index=False))
print("\n" + "="*80)
print("✓ Checkpoint 0.1 Complete: All 3 datasets loaded and inspected.")
print("Next: Phase 0 Step 0.2 - EDA & Cleaning Pipeline")
print("="*80)
```
